# Rental Property Occupancy Analysis

This notebook analyzes the scraped availability data from weekendica.com properties.

```python
import pandas as pd
from pathlib import Path

EXCEL_FILE = Path('KuceZaIzdavanje.xlsx')
```

## 1. Check Excel Structure

```python
# Read both sheets
props = pd.read_excel(EXCEL_FILE, sheet_name='Properties')
avail = pd.read_excel(EXCEL_FILE, sheet_name='Availability')

print('=== PROPERTIES SHEET ===')
print(f'Shape: {props.shape} (rows, columns)')
print(f'\nAll columns ({len(props.columns)} total):')
for i, col in enumerate(props.columns):
    print(f'  {i+1}. {col}')
```

```python
print('=== AVAILABILITY SHEET ===')
print(f'Shape: {avail.shape}')
print(f'\nColumns: {list(avail.columns)}')
print(f'\nSample data:')
avail.head(6)
```

## 2. View Analytics Columns (Properties Sheet)

```python
# Find analytics columns
analytics_cols = [c for c in props.columns if c.startswith('Occ_') or c == 'Rank']
print(f'Analytics columns found: {analytics_cols}')
print(f'\nThese are at columns: {[props.columns.get_loc(c)+1 for c in analytics_cols]}')
```

```python
# Show properties with analytics
display_cols = ['ID', 'Lokacija'] + analytics_cols
props[display_cols]
```

## 3. Occupancy Summary

```python
# Overall stats
print('=== AVAILABILITY STATS ===')
print(f'Total records: {len(avail)}')
print(f'Unique properties: {avail["property_id"].nunique()}')
print(f'Date range: {avail["date"].min()} to {avail["date"].max()}')
print(f'\nBooked vs Available:')
print(avail['booked'].value_counts().rename({0: 'Available', 1: 'Booked'}))
```
